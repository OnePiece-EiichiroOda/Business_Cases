```python
males_using_KP281 = df[(df["Product"]=="KP281") & (df["Gender"]=="Male")].shape[0]
males_using_KP281
```

```python
females_using_KP281 = df[(df["Product"]=="KP281") & (df["Gender"]=="Female")].shape[0]
females_using_KP281
```

```python
males_using_KP481 = df[(df["Product"]=="KP481") & (df["Gender"]=="Male")].shape[0]
males_using_KP481
```

```python
females_using_KP481 = df[(df["Product"]=="KP481") & (df["Gender"]=="Female")].shape[0]
females_using_KP481
```

```python
males_using_KP781 = df[(df["Product"]=="KP781") & (df["Gender"]=="Male")].shape[0]
males_using_KP781
```

```python
females_using_KP781 = df[(df["Product"]=="KP781") & (df["Gender"]=="Female")].shape[0]
females_using_KP781
```

```python
plt.figure(figsize=(12, 5))
data_plot = {
 'Product': ['KP281','KP481','KP781'],
 'Male': [males_using_KP281,males_using_KP481,males_using_KP781],
 'Female': [females_using_KP281,females_using_KP481,females_using_KP781]
}
# Convert data to a pandas DataFrame
df_plot= pd.DataFrame(data_plot)
# Melt the DataFrame to convert it to a "long" format
df_melted = df_plot.melt('Product', var_name='Group', value_name='Value')
# Set the style
sns.set(style='whitegrid')
# Plot the grouped bar chart
sns.barplot(x='Product', y='Value', hue='Group', data=df_melted)
# Set labels and title
plt.xlabel('Products')
plt.ylabel('Values')
plt.title('Grouped Bar Chart')
# Display the plot
plt.show()
```

We can see that both male and female are using similar range of KP281 and almost same with KP481 and not so equal at KP781.

Clearly we can say that there is male dominance on each category of products.

```python
plt.figure(figsize=(12, 5))
# Create a figure with two subplots side by side
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))
# Plot movies on the first subplot
sns.lineplot(x=male_sorted_age, y=male_age_count, marker='o', markersize=8, ax=axes[0])
axes[0].set_title("Number of Males with age")
axes[0].set_xlabel("Age")
axes[0].set_ylabel("Count")
# Plot TV shows on the second subplot
sns.lineplot(x=female_sorted_age, y=female_age_count, marker='o', markersize=8, ax=axes[1])
axes[1].set_title("Number of Females with age")
axes[1].set_xlabel("Age")
axes[1].set_ylabel("Count")
# Adjust the spacing between subplots
plt.tight_layout()
# Display the figure
plt.show()
```

**Insight for Males** : We can see that males from the age 20-26 there are maximum numbers who are interested in fitness, But then from age 26-45 it just decresed way down and fluctuates showing less numbers in fitness.

**Insight for Females** : We can see that females from the age 20-25 there are maximum numbers who are interested in fitness, But then from age 25-40 it just decresed way down and fluctuates showing less numbers in fitness.

Numbers for Males at any age are greater than females, showing that there are  males are more into fitness than females.

```python
#Preparing the data for visualisation
def custom_agg(x):
 return x.nunique()
data = pd.pivot_table(df, index = male['Age'],
 columns = male['Fitness'],values='Product', aggfunc=custom_agg)
```

```python
sns.set(style='white', palette='viridis')
# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data, annot=True, fmt=".0f", cmap='viridis', cbar=True)
# Set labels and title
plt.xlabel('Fitness')
plt.ylabel('Age')
plt.title('Heatmap of Product count by Age and Fitness Of Males')
# Adjust the spacing
plt.tight_layout()
# Display the plot
```

Above Heatmap shows us the count of products used by Males at all the given age groups based on theie fitness levels.

It shows that maximum males who are level 3 at fitness are using most of the products and with level 5 fitness at second most and following with level 4, level 2, level 1.

```python
#Preparing the data for visualisation
def custom_agg(x):
 return x.nunique()
data = pd.pivot_table(df, index = female['Age'],
 columns = female['Fitness'],values='Product', aggfunc=custom_agg)

sns.set(style='white', palette='viridis')
# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data, annot=True, fmt=".0f", cmap='viridis', cbar=True)
# Set labels and title
plt.xlabel('Fitness')
plt.ylabel('Age')
plt.title('Heatmap of Product count by Age and Fitness Of Females')
# Adjust the spacing
plt.tight_layout()
# Display the plot
```

Above Heatmap shows us the count of products used by Females at all the given age groups based on theie fitness levels.

It shows that maximum females who are level 3 at fitness are using most of the products and with level 2 fitness at second most and following with level 4, level 5, level 1.

```python
data = pd.pivot_table(df, index = 'Miles',
 columns = 'Product',values='Usage')

plt.figure(figsize=(10, 8))
sns.heatmap(data, annot=True, fmt=".0f", cmap='viridis', cbar=True)
# Set labels and title
plt.xlabel('Product')
plt.ylabel('Miles')
plt.title('Usage-wise Miles completed of Product')
# Adjust the spacing
plt.tight_layout()
# Display the plot
plt.show()
```

1. There are more people who used KP281 and completed almost low to high range of
miles in that product with Usage band of 2-5.

2. There who used KP481 and completed almost low to high range of miles in that product with Usage band of 2-5 are similar to KP281.

3. But people who used KP781 and completed miles from mid range to Highest range in that product with Usage band of 4-7.

***We conclude that people are more confident in using K781 to fullest capacity as it is the highest quality product. And KP281 is used by most people because it  achieves basic human capacity fitness.***
